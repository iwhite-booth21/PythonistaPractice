```python
# Histplot is a access level plot

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

a_data = pd.read_csv('Datasets/Attrition.csv')

f, ax = plt.subplots(figsize = (12,6))
sns.histplot(data = a_data, x = "MonthlyIncome")
```

```python
sns.displot(data = a_data, x = "TotalWorkingYears", aspect = 2)
```

```python
sns.displot(data = a_data, x = "TotalWorkingYears", binwidth =5, aspect = 2)
```

```python
sns.displot(data = a_data, x = "NumCompaniesWorked", aspect = 2)
```

```python
sns.displot(data = a_data, x = "NumCompaniesWorked", discrete = True, aspect = 2)
```

```python
sns.displot(data = a_data, x = "NumCompaniesWorked", discrete = True, shrink = 0.5, aspect = 2)
```

```python
#Kernal Density Estimation Curve
#KDE is a smoothing function
# The bandwith can be tweaked to get rid of the non probabilities for extreme values
sns.displot(data = a_data, x = "MonthlyIncome", kind = 'kde', aspect = 2)
```

```python
# For the Ages
# All values must sum up to 1
# Can be use to get a approximation of the probability distribution function

f , ax = plt.subplots(figsize = (12, 6))
sns.kdeplot(data = a_data, x = "Age")
```

```python
# Creating Univirate Plots
# The KDE is present on the histogram
sns.displot(data=a_data, x = "MonthlyIncome", kde = True, aspect = 2)
```

```python
# Creating Univirate Plots
# The KDE is present on the histogram
# Rug gives specific bin data
sns.displot(data=a_data, x = "MonthlyIncome", kde = True, rug = True, aspect = 2)
```

```python
# Using access level plots
# Using rugplot gives us specific data about the bins 
f, ax = plt.subplots(figsize = (12,6))
sns.histplot(data = a_data, x = "Age", kde = True)
sns.rugplot(data = a_data, x = "Age",)
```

```python
# Lets examine the role of the bandwidth
f, ax = plt.subplots(figsize = (12, 6))
# Use the bandwidth param, and list it in the label
# The bandwidth is a multiplicative factive of the bandwidth
# This multiplies the smoothing of the standard deviation of the smoothing kernal
# BW too large - Over smoothing
# BW too small - Too small/under-smoothing "Noise
"
sns.kdeplot(data = a_data, x = "HourlyRate", bw_adjust = 0.05, label = "bw: 0.05"); # Smoothing Bandwidth - Really Small - Very Jagged
sns.kdeplot(data = a_data, x = "HourlyRate", bw_adjust = .2, label = "bw: .2"); # Smoothing Bandwidth - Intermediate - Less Jagged
sns.kdeplot(data = a_data, x = "HourlyRate", bw_adjust = 2, label = "bw: 2"); # Smoothing Bandwidth - Normal - Smooth
plt.legend()
```

```python
# Incorparating cut param - Removes extreme vakues from kde plots
sns.kdeplot(data = a_data, x = "HourlyRate", bw_adjust = 0.05, cut = 0, label = "bw: 0.05"); # Smoothing Bandwidth - Really Small - Very Jagged
sns.kdeplot(data = a_data, x = "HourlyRate", bw_adjust = .2, cut = 0, label = "bw: .2"); # Smoothing Bandwidth - Intermediate - Less Jagged
sns.kdeplot(data = a_data, x = "HourlyRate", bw_adjust = 2, cut = 0, label = "bw: 2"); # Smoothing Bandwidth - Normal - Smooth
plt.legend()
```
