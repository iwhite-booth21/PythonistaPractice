```python
import pandas as pd
import numpy as mp 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# /telecomm data
t_data = pd.read_csv("Datasets/telecom_data_churn.csv")
t_data.head()
```

```python
# Tranpose the csv data using describe method and .T
t_data.describe().T
```

```python
# Using catplot to get a box and whiskerplot - using y
sns.catplot(y = "Total day charge", kind = 'box', data = t_data, height = 6)
```

```python
# Using catplot to get a box and whiskerplot - using x
sns.catplot(x = "Total day charge", kind = 'box', data = t_data, aspect = 2)
```

```python
# Using catplot to get a box and whiskerplot - using access level plot
f, ax = plt.subplots(figsize = (12, 7)) # Access level plot
sns.boxplot(y = "Total day calls", data = t_data)
```

```python
# We will add in the Churn param for the boxplot
# Left boxplot is churn of 0 and right boxplot is 1
sns.catplot(x = "Churn", y = "Total day charge", kind = "box", data = t_data, aspect = 2);
```
