```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

Dataset link: https://www.kaggle.com/pavansubhasht/ibm-hr-analytics-attrition-dataset?select=WA_Fn-UseC_-HR-Employee-Attrition.csv

```python
attrition_data = pd.read_csv('Datasets/Attrition.csv')

attrition_data.head()
```

A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color (analagous to a heatmap()). Similarly, a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian. The default representation then shows the contours of the 2D density.
Here, We have drawn a bivariate histograms of 'Age' and 'MonthlyIncome' using displot function, which is Figure level function built on top of FacetGrid. The highest count of observations can be seen around 28 of Age and 2500 of Annual Income.

```python
sns.displot(data = attrition_data , x = 'Age', y = 'MonthlyIncome', aspect = 2)
```

Same plot can be drawn using 'histplot' function,which is a axes level function

```python
f, ax = plt.subplots(figsize = (12, 6))

sns.histplot(data = attrition_data , x = 'Age', y = 'DistanceFromHome')
```

We have plotted bivariate KDE curves of 'PercentSalaryHike' and 'MonthlyIncome' by setting 'kind' to 'kde'.
The highest densities can be observed at around 12.5 for percentSalary hike and corresponding values of Monthly Income are at around 4000 and 18000.

```python
sns.displot(data = attrition_data ,  x = 'PercentSalaryHike', y = 'MonthlyIncome', 
            kind = 'kde', aspect = 2)
```

We have added another variable 'Attrition' as hue,which differentiates based on colors.This works best when there is minimum overlapping between the conditional distributions.
The curves for each level of 'Attrition' are  different. The density for 'No' curve has another peak at around 1700 of MonthlyIncome and 12.5 percent Salary Hike,which is not the case with 'Yes' curve.

```python
sns.displot(data = attrition_data ,  x = 'PercentSalaryHike', 
            y = 'MonthlyIncome', hue = "Attrition",
            kind = 'kde', aspect = 3)
```

Here, We have drawn a bivariate histograms of 'DailyRate' and 'MonthlyIncome' using displot function, We have enabled the color bar to show the mapping between counts and color intensity.
The highest count of observations at around 35 can be seen at multiple rectangles at around 1300-1400 ,800-900 of 'Daily Rate' and 2500 of Monthly Income.

```python
sns.displot(data = attrition_data , x = 'DailyRate', y = 'MonthlyIncome',
            cbar = True, aspect = 2)
```

The no. of bins for the both distributions can be customised by passing a tuple to 'bins' argument.We seem to have increased the no. of bins as gaps are visible.The Shapes and color intensity may vary with the setting of bins.

```python
sns.displot(data = attrition_data , x = 'DailyRate', y = 'MonthlyIncome', 
            cbar = True, bins = (20, 40), aspect = 2)
```

Here, We have drawn a bivariate histograms of 'MonthlyRate' and 'MonthlyIncome' using displot function,

The highest count of observations at around 35 can be seen at multiple rectangles centred at 5000 and 7000 of 'Daily Rate' and centred at  2500 of Monthly Income.

```python
sns.displot(data = attrition_data , x = 'MonthlyRate', y = 'MonthlyIncome',
            cbar = True, aspect = 2)
```

We have added 'Attrition' as hue.It works best when there is minimal overlap of conditional distributions.
2 rectangles in lower extreme and 2 in upper extreme of Monthly Rate for Monthly income in the range 1250 to 2000 for employees who have left organisation(Attrition=Yes).
The highest counts for Employees who have not left matches with observations from the above plot without hue.

```python
sns.displot(data = attrition_data , x = 'MonthlyRate', y = 'MonthlyIncome',
            cbar = True, hue = 'Attrition', aspect = 2)
```

We have plotted two discrete variables 'Department' and 'Attrition' which shows the cross-tabulation of the observations. We have enabled the color bar to show the mapping between counts and color intensity.
It can be observed that both  highest no. of Attritions happened in R&D department.Also,The highest no. of employees have not left the organisation are in R&D department only.

```python
sns.displot(data = attrition_data , x = 'Department',
            y = 'Attrition' , cbar = True, aspect = 2)
```

Here, We observed that the output is Joint Grid Object.So jointplot function is an figure level function. A figure-level function has complete control over the entire figure and each time a figure level function is called, it creates a new figure which can include multiple Axes, all organised in a meaningful way. This is a scatter plot along  with the univariate distributions in the form of histogram of the 'YearsAtCompany' as x and 'MonthlyIncome' as y variables. It can be observed through the scatter plot that with the increase in 'YearsAtCompany', the 'MonthlyIncome' also increases.

```python
sns.jointplot(x = 'YearsAtCompany', y = 'MonthlyIncome', 
              data = attrition_data, height = 8)
```

Here in the below plot, We have added Regression line to the plot using 'kind'='reg. Clearly Positive correlation can be observed between the variables.Also, By default dropna is False,we have set dropna to True to remove any cases of missing x and y values.

```python
sns.jointplot(x = 'YearsAtCompany', y = 'MonthlyIncome', 
              kind = 'reg', dropna = True, 
              data = attrition_data, height = 8)
```

Here is the Hexbin plot represents the relationship of two continuos variables 'Age ' and 'YearsAtCompany'. It is particularly used when you have a lot of data points. Instead of overlapping, the plotting window is split in several hexbins, and the number of points per hexbin is counted. The density of the color denotes this number of points.We have set the color to Green using color='g'. So in the below plot the darkest bins for the plot are at 28,32,34 for 'Age' and is around 2.5 for 'YearsAtCompany'.

```python
sns.jointplot(x = 'Age' , y = 'YearsAtCompany', kind = 'hex', 
              color = 'g', data = attrition_data, height = 8)
```

We can customise  the limits on X and Y axes using xlim and ylim arguments for getting magnified view for a particular range of values in both axes.

```python
sns.jointplot(x = 'Age' , y = 'YearsAtCompany', kind = 'hex', 
              color = 'g', data = attrition_data, xlim = (25, 40), 
              ylim = (0, 15), height = 8)
```

We have plotted KDE curves instead of scatterplots(Joint axes) and histograms( in marginal axes) by setting 'kind' to 'kde'. Multimodal peaks( at 3 and 4 for 'Performance rating' and 4000 and 17000 for 'Monthly income') can be observed in both marginal axes and is supported by the contors on the joint axes.

```python
sns.jointplot(x = 'PerformanceRating', y = 'MonthlyIncome', kind = 'kde',
              data = attrition_data, height = 10)
```

A easier way to show marginal distributions uses a “rug” plot, which adds a small tick on the edge of the plot to represent each individual observation.We have set 'rug' to True to draw marginal rug plots.

```python
sns.displot(data = attrition_data , x = 'PerformanceRating', y = 'MonthlyIncome', 
            kind = 'kde', rug = True, aspect = 2)
```

Same above plot can be drawn using axes level functions kdeplot and rugplot, only the tick colors by default is different which can be customised.

```python
f, ax = plt.subplots(figsize = (12, 10))

sns.kdeplot(data = attrition_data ,  x = 'PerformanceRating', y = 'MonthlyIncome') 

sns.rugplot(data = attrition_data ,  x = 'PerformanceRating', y = 'MonthlyIncome' )            
```
