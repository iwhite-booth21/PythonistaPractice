```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

Dataset link: https://www.kaggle.com/pavansubhasht/ibm-hr-analytics-attrition-dataset?select=WA_Fn-UseC_-HR-Employee-Attrition.csv

```python
attrition_data = pd.read_csv('Datasets/Attrition.csv')

attrition_data.head()
```

```python
attrition_data.describe().T
```

First of all, we observed that the output heading is a FacetGrid object, which means it is an Figure level function. Each plotting function in seaborn is either a figure-level function or an Axes-level function. An Axes-level function draws onto a single matplotlib Axes and does not effect the rest of the figure. A figure-level function, on the other hand, controls the entire figure. A figure-level function can call different Axes-level functions to draw different types of subplots on different Axes.
Below fig shows  an univariate distribution of 'Age' in seaborn using the displot() function. By default, this will draw a histogram by using axes level histplot function with Counts on Y-axis. Here it is observed that the most frequent observation is at around 28.

```python
sns.displot(data = attrition_data, x = 'Age', aspect = 2)
```

Same plot can be drawn by using axes level histplot function.

```python
f, ax = plt.subplots(figsize = (12, 6))

sns.histplot(data = attrition_data, x = 'MonthlyIncome')
```

We can set the binwidth and  number of bins by passing values to the 'bins' and 'binwidth' argument. Histograms shape varies with the number of  bins selected. Also note that We can hide the Output heading by using ';' after the code of the plot.

```python
sns.displot(data = attrition_data, x = 'TotalWorkingYears', aspect = 2)
```

```python
sns.displot(data = attrition_data, x = 'TotalWorkingYears', binwidth = 5, aspect = 2)
```

In case of variable taking discrete integer values,There are some gaps visible in the plots.

```python
sns.displot(data = attrition_data, x = 'NumCompaniesWorked', aspect = 2)
```

This gap can be removed by setting discrete=True, which chooses bin breaks that represent the unique values in a dataset with bars that are centered on their corresponding value.

```python
sns.displot(data = attrition_data, x = 'NumCompaniesWorked',
            discrete = True, aspect = 2);
```

Here,The categorical attribute of the variable is highlighted using 'shrink', the histogram bars are centred at unique values and bars are shrinked.

```python
sns.displot(data = attrition_data, x = 'NumCompaniesWorked', 
            discrete = True, 
            shrink = 0.5, aspect = 2);
```

Kde curve can be drawn using kind='kde'. The peak of the distribution is at around 3500 of Monthly Income.

```python
sns.displot(data = attrition_data, x = 'MonthlyIncome', kind = 'kde' , aspect = 2)
```

Same KDE plot can be drawn using axes level kdeplot function.

```python
f, ax = plt.subplots(figsize = (12, 6))

sns.kdeplot(data = attrition_data, x = 'Age')
```

Both histogram and KDE curves can be drawn together by setting kde=True inside displot function.Nature of distribution can be better analysed using this approach.

```python
sns.displot(data = attrition_data, x = 'MonthlyIncome', 
            kde = True , aspect = 2)
```

On the same plot, We have drawn a a Rug plot showing distribution of 'Monthly Income', which draws a small vertical tick at each observation(unique occurences) . You can make the rug plot itself with the rugplot() function, but it is also available in distplot(). It can be observed that observations are crowded in between 2000 and 7500.

```python
sns.displot(data = attrition_data, x = 'MonthlyIncome', 
            kde = True , rug = True, aspect = 2)
```

Same plot can be achieved using axes level functions

```python
f, ax = plt.subplots(figsize = (12, 6))

sns.histplot(data = attrition_data, x = 'Age' , kde = True)

sns.rugplot(data = attrition_data, x = 'Age')
```

Here, We have drawn KDE curves for 'Hourly Rate' with different bandwidth values. Smoothness of the curve increases with increase in bw_adjust values.

```python
f, ax = plt.subplots(figsize = (12, 6))

sns.kdeplot(data = attrition_data, x = 'HourlyRate' ,
            bw_adjust= 0.05, label = 'bw: 0.05');

sns.kdeplot(data = attrition_data, x = 'HourlyRate', 
            bw_adjust = .2, label = 'bw: 0.2');

sns.kdeplot(data = attrition_data, x = 'HourlyRate', 
            bw_adjust = 2, label = 'bw: 2');

plt.legend()
```

The overshooting of KDE curves beyond the lower and upper limit can be avoided by using cut=0.

```python
f, ax = plt.subplots(figsize = (12, 6))

sns.kdeplot(data = attrition_data, x = 'HourlyRate', 
            bw_adjust= 0.05, cut = 0, label = 'bw: 0.05');

sns.kdeplot(data = attrition_data, x = 'HourlyRate', 
            bw_adjust = .2, cut = 0, label = 'bw: 0.2');

sns.kdeplot(data = attrition_data, x = 'HourlyRate', bw_adjust = 2, 
            cut = 0, label = 'bw: 2');

plt.legend()
```

Here, We have drawn a histogram for 'PercentSalaryHike' and 'PerformanceRating' as  hue variable plotted separate histogram for each of its levels and levels are differentiated based on colors. We have customised hue palette to 'deep'.
It can be observed that distributions are separate for different levels of Performance Rating.
For high level i.e 4  of Performance rating,the distribution is centred towards higher Values of 'PercentSalaryHike'.

```python
sns.displot(data = attrition_data, x = 'PercentSalaryHike', 
            hue = 'PerformanceRating', palette = 'deep' , aspect = 2)
```

Instead of layering each bar, they can be â€œstackedâ€ or moved vertically(dodged). Here, We have stacked multiple hue levels.It can be observed that in most of bins ,major chunk is comimg from Research and Development Department.

```python
sns.displot(data = attrition_data, x = 'MonthlyIncome', 
            hue = 'Department', multiple = 'stack', aspect = 2)
```

Here we have used 'dodge' option ,which moves them horizontally and reduces their width. No overlapping is observed and the bars remain comparable in terms of height. But it only works well when the categorical variable has a small number of levels

```python
sns.displot(data = attrition_data, x = 'MonthlyIncome', 
            hue = 'Department', multiple = 'dodge', aspect = 2)
```

Here, We have drawn distribution for 'PercentSalaryHike' and added 'Department'  using col,which facets through columns.For each level of 'Department',distribution for 'PercentSalaryHike' is drawn. The peak is at  14 for 'Sales' and 13 for  'R&D' Department. For Hr dept, peak is at around 11.

```python
sns.displot(data = attrition_data, x = 'PercentSalaryHike', 
            col = 'Department' , discrete = True );
```

When the subsets have unequal numbers of observations, comparing their distributions in terms of counts may not be ideal. One solution is to normalize the counts using the stat parameter.Here, We have drawn histogram for 'Years At Company' and added 'Department' as hue,which differentiates levels based on colors.Peak of histogram bars is at 5 of 'Years At company' with major chunk of bar is contributed by R&D department.

```python
sns.displot(data = attrition_data, x = 'YearsAtCompany', discrete = True, 
            multiple = 'dodge', hue = 'Department' , aspect = 2);
```

Here, We have drawn KDE curve for 'Age' and introduced another variable 'Attrition' using hue. It can be observed that peak of the curves for each level of 'Attrition' is different. For employee Attrition to be Yes,the peak is at around 30 of Age.For employee Attrition to be No ,the peak is at around 36 of Age.

```python
sns.displot(data = attrition_data, x = 'Age', hue = 'Attrition', 
            kind = 'kde' , aspect = 2)
```

It is also possible to fill in the curves for single or layered densities, although the default alpha value (opacity) will be different, so that the individual densities are easier to resolve.

```python
sns.displot(data = attrition_data, x = 'Age', hue = 'Attrition', 
            kind = 'kde', fill = True, aspect = 2)
```

A third option for visualizing distributions computes the â€œempirical cumulative distribution functionâ€ (ECDF). This plot draws a monotonically-increasing curve through each datapoint such that the height of the curve reflects the proportion of observations with a smaller value. So according to the plot,around 16-17% employees have got less than 12% Salary Hike.

```python
sns.displot(data = attrition_data, x = 'PercentSalaryHike', kind = 'ecdf');
```

Multiple distributions can be compared using hue.

```python
sns.displot(data = attrition_data, x = 'PercentSalaryHike',
            hue = 'Attrition', kind = 'ecdf');
```
