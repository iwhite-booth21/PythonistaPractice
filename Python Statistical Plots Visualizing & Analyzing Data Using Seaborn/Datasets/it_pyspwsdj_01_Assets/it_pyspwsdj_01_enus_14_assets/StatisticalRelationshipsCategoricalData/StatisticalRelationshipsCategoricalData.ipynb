```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

Dataset link : https://www.kaggle.com/mnassrib/telecom-churn-datasets?select=churn-bigml-20.csv

```python
telecomm_data = pd.read_csv('Datasets/telecom_data_churn.csv')

telecomm_data.head()
```

Viewing descriptive statistics of the dataset.

```python
telecomm_data.describe().T
```

We have drawn a box plot using Catplot function by using kind = 'box' to show the distribution of 'Total day charge'. .The Median value of Total day charge is at around 31,which matches the median value of 30.31 obtained in Descriptive statistics earlier.Two outliers are visible towards upper extreme and one towards lower extreme.Note that the boxplot is plotted horizontally as we have passed the variable in 'x' argument.

```python
sns.catplot(y = 'Total day charge', kind = 'box', data = telecomm_data, height = 6)
```

Here, We have drawn a box plot to show distribution for 'Total eve charge'.The Median value of Total eve charge is at around 17,which matches the median value of 17.31 obtained in Descriptive statistics earlier.Few outliers are visible towards both extremes.Note that the boxplot is plotted vertically as we have passed the variable in 'y' argument.

```python
sns.catplot(x = 'Total eve charge', kind = 'box', data = telecomm_data , aspect = 2)
```

Same box plot can be drawn by using boxplot function, which is an axes level function.
The Median value of Total night charge is at around 9,which matches the median value of 9.07 obtained in Descriptive statistics earlier.Few outliers are visible towards both extremes.

```python
f, ax = plt.subplots(figsize = (12, 8))

sns.boxplot(y = 'Total day calls', data = telecomm_data )
```

We have drawn a box plot using Catplot function by  using kind='box' to show the distribution or spread of 'Total day charge' along the Y-axis for each level of 'Churn' in X axis. The distribution or spread is same for both levels of 'Churn'.
But the Median value of Total day charge for Churned customers is higher than Non-churned customers and there are some outliers visible in Non-churned category. Also, the interquartile range is higher for Churned category.



```python
sns.catplot(x = 'Churn', y = 'Total day charge', 
            kind = 'box', data = telecomm_data , aspect = 2);
```

We have used boxplot function to draw the same boxplot. Note that boxplot function is a axes level function unlike Catplot function.

```python
f, ax = plt.subplots(figsize = (12, 8))

sns.boxplot(x = telecomm_data['Churn'], y = telecomm_data['Total day calls'])
```

Here, We have plotted Box plot of categorical variable 'International plan'(X-axis) with 'Total international charge' on Y-axis.
The median value of international charges are almost same for both set of customers which are having an international plan and those not having an international plan. The international charges values are more spread for those customers not having international plan and no. of outliers are also more.

```python
f, ax = plt.subplots(figsize = (12, 8))

sns.boxplot(x = 'International plan', y = 'Total intl charge', data = telecomm_data);
```

Display of  Outliers can be disabled by setting 'showfliers' to False

```python
f, ax = plt.subplots(figsize = (12, 8))

sns.boxplot(x = 'International plan', y = 'Total intl charge', 
            showfliers = False, data = telecomm_data)
```

We have added one more variable 'Churn ' to the plot by using hue,which will differentiate levels based on color.
We have customised the palette('Set2') for 'hue' variable.It  can observed that those churned customers,with an international plan, have a median value of international charges marginally higher than others. It can also be observed that inter quartile range for 'Total international charge' of  Non-churned customers with no international plan is highest among all other categories.

```python
f, ax = plt.subplots(figsize = (12, 8))

sns.boxplot(x = 'International plan', y = 'Total intl charge', hue = 'Churn', 
            palette = 'Set2', data = telecomm_data);
```

When we set dodge to False, the two levels of hue are jumbled together and the distributions are not clear.

```python
f, ax = plt.subplots(figsize = (12, 8))

sns.boxplot(x = 'International plan',y = 'Total intl charge', 
            hue = 'Churn', dodge = False, 
            palette = 'Set2', data = telecomm_data);
```

Here, We have plotted Box plot of categorical variable 'Churn'(X-axis) with 'Total night charge' on Y-axis.We have added one more variable 'International plan' to the plot by using hue.We have customised the palette and increased the line width using 'linewidth' argument.It can be observed that inter quartile range for 'Total night charge' of  Non-churned customers with no international plan is highest among all other categories.Also, the median value of night charge for customers with no international plan is higher than customers with international plan for both Churned and Non-churned categories.


```python
f, ax = plt.subplots(figsize = (12, 8))

sns.boxplot(x = 'Churn',y = 'Total intl minutes', hue = 'International plan' , 
            palette = 'Set2', linewidth = 3, data = telecomm_data);
```

We can also draw a boxen  plot of categorical variable 'International plan'(X-axis) with 'Total eve charge' on Y-axis  which is  similar to box plot but it is more detailed with more number of quantiles.The median for both categories of 'International plan' are same at around 17.The distribution of 'Total eve charge' for the customers with no international plan  is having more spread than customers with international plan.


```python
sns.catplot(x = 'International plan',y = 'Total intl charge', 
            kind = 'boxen' , data = telecomm_data , aspect = 3);
```

We can draw the same boxenplot using using boxenplot function,which is an axes level function.

```python
f, ax = plt.subplots(figsize = (12, 8))

sns.boxenplot(x = 'International plan', y = 'Total intl calls', 
              data = telecomm_data ) 
```

We can add more variables by faceting through both columns and rows. Here we have added 'Voice mail plan' variable with two levels of 'Voice mail plan'  faceting through columns.
The median value of International charge for Churned customers with International plan and no voicemail plan is highest among all. The Inter Quartile Range  is highest for Non-churned customers with no international plan for both type of customers having a voicemail plan and customers not having a voice mail plan.

```python
sns.catplot(x = 'International plan', y = 'Total intl charge', hue = 'Churn',
            col = 'Voice mail plan', kind = 'box', data = telecomm_data );
```

```python
sns.catplot(y = 'Number vmail messages', kind = 'violin', data = telecomm_data , aspect = 2);
```

```python
sns.catplot(x = 'Customer service calls', kind = 'violin', data = telecomm_data  , aspect = 3);
```

For the below plot We have taken  'Churn' as the categorical variable in X-axis and 'Total eve charge' in Y-axis. It is observed that median value of evening  charge( around 20) for churned customers is higher than non-churned customers(around 17)

```python
sns.catplot(x = 'Churn', y = 'Number vmail messages', 
            kind = 'violin', data = telecomm_data, aspect = 3);
```


Same plot can be drawn by using violinplot function,which is an axes level function like Box plot function . Scale argument controls the width of violins. Please note that scale argument is by default set to area, so the area enclosed by each violins is same

```python
f, ax = plt.subplots(figsize = (12, 8))

sns.violinplot(x = 'Churn', y = 'Customer service calls', data = telecomm_data)
```

```python
f, ax = plt.subplots(figsize = (12, 8))

sns.violinplot(x = 'Churn', y = 'Total intl charge', 
               data = telecomm_data);
```

For the below plot We have taken  'Churn' as the categorical variable in X-axis and 'Total intl charge' in Y-axis.If 'scale' argument is set to count,then width of the violins will depend on  the number of observations in that bin. So it is clearly visible that number of churned customers is very less compared to non-churned customers.



```python
f, ax = plt.subplots(figsize = (12, 8))

sns.violinplot(x = 'Churn', y = 'Total intl charge', 
               scale = 'count', data = telecomm_data);
```

If 'scale' argument is set to width,then width of all the violins will be same.


```python
f, ax = plt.subplots(figsize = (12, 8))

sns.violinplot(x = 'Churn', y = 'Total intl charge', 
               scale = 'width', data = telecomm_data);
```

Here,We have plotted a violin plot with 'Churn' as categorical variable and 'Total intl calls' as Y variable.
We have added the variable 'International plan' as hue,which differentiates levels based on colors.
Also, the spread of distribution for non-churned customers with no international plan is higher than other categories.
For churned customers,median of international calls is same for both type customers with international plan and customers not with international plan but customers with no international have distribution wider than that of customers with international plan.
For non-churned customers, median of international calls is slightly higher for customers with international plan.

```python
sns.catplot(x = 'Churn', y = 'Total intl calls', hue = 'International plan',
            kind = 'violin', data = telecomm_data , aspect = 2)
```

Here,We have plotted a violin plot with 'Churn' as categorical variable and 'Total intl calls' as Y variable .Titles for each plot with Bw values are shown using matplotlib pyplot 'title' method.
It can be observed that smoothness of the curves is increasing with increase in 'bw' values. Value for cut argument is set to zero to limit the curves upto minimum value. For lower bw value i.e 0.2, the multimodal nature of curves is visible.

```python

sns.catplot(x = 'Churn', y = 'Total intl calls', hue = 'International plan',
            kind = 'violin', bw = 0.2, cut = 0, data = telecomm_data, aspect = 2)

plt.title('Bw:0.2')

sns.catplot(x = 'Churn', y = 'Total intl calls', hue = 'International plan',
            kind = 'violin', bw = 0.5, cut = 0, data = telecomm_data, aspect = 2)

plt.title('Bw:0.5')


sns.catplot(x = 'Churn', y = 'Total intl calls', hue = 'International plan',
            kind = 'violin', bw = 1, cut = 0, data = telecomm_data, aspect = 2)

plt.title('Bw:1')

sns.catplot(x = 'Churn', y = 'Total intl calls', hue = 'International plan',
            kind = 'violin', bw = 4, cut = 0, data = telecomm_data, aspect = 2)

plt.title('Bw:4')

```

Here,We have plotted a violin plot with 'Churn' as categorical variable and 'Total night minutes' as Y variable.
We have added variable  'Voice mail plan' by using col which facets through columns.
For both type of customers with voice mail plan and without voice mail plan,the median of 'Total night minutes' is slightly higher for Churned customers than non-churned category and the distribution is slightly wider for Non-churned category.

```python
sns.catplot(x = 'Churn', y = 'Total day minutes' , col =  'Voice mail plan',
            kind = 'violin', data = telecomm_data , aspect = 1.5)
```

Here, We have drawn a violin plot with  'Churn' as Categorical variable and 'Total intl charge' as Y variable. We have also added variable 'International plan' as hue. We can also split the violins when the hue parameter has only two levels for efficient space utilisation. Please note that median value for each level of hue are not shown separately and the median value shown is nothing but the median for that level of categorical variable in X-axis.

```python
sns.catplot(x = 'Churn', y = 'Total day minutes' , hue =  'Voice mail plan',
            kind = 'violin', split = True, data = telecomm_data ,aspect = 1.5)
```

For the same plot, We are showing  each individual observation instead of the summary boxplot values by passing 'stick' to inner argument.We have set the hue palette to 'Set2'.
So, As expected the concentration of sticks are more near the peaks of the distribution.

```python
sns.catplot(x = 'Churn', y = 'Total day minutes' , hue =  'Voice mail plan',
            kind = 'violin', inner = 'stick', palette = 'Set2',
            split = True, data = telecomm_data , aspect = 3);
```

For the same plot, quartiles of the distribution can be drawn using inner='quartile'

```python
sns.catplot(x = 'Churn', y = 'Total day minutes' , hue =  'Voice mail plan',
            kind = 'violin', inner = 'quartiles', palette = 'Set2',
            split = True, data = telecomm_data , aspect = 3);
```

Here, We have plotted 'Churn' as Categorical variable and 'Total day charge' as Y variable.
We have combined the violin plot and Swarm plot and all the observations are shown with distribution.
The distributions of data points for 'Total day charge'  for both levels  as depicted inside the Violin plot in the form of swarm plot are also different. The density estimate curves of 'Total day charge' for both levels are also different with two peaks visible for 'Churned' category and distribution is wider for the same category.

```python
g = sns.catplot(x = 'Churn', y = 'Total day charge', kind = 'violin', 
                inner = None, data = telecomm_data, aspect = 2)

sns.swarmplot(x = 'Churn', y = 'Total day charge', color = 'k', 
              size = 3, data = telecomm_data, ax = g.ax)
```
