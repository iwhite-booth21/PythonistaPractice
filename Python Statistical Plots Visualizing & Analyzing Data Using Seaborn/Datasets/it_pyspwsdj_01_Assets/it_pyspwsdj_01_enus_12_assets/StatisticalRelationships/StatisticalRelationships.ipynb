```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

Here, We have loaded the dataset.
dataset link :https://www.kaggle.com/zaurbegiev/my-dataset?select=credit_train.csv


```python
credit_data = pd.read_csv('Datasets/credit_train.csv')

credit_data.head()
```

```python
credit_data.shape
```

```python
credit_data = credit_data.drop_duplicates()

credit_data.shape
```

We have dropped all  cases having NAs using 'dropna' method.

```python
credit_data = credit_data.dropna()

credit_data.shape
```

Viewing the Descriptive statistics of the Dataset

```python
credit_data.describe().T
```

Above plot is a  scatter plot between 'Annual' and 'Monthly Debt' which is the default kind of plot in relplot. It is observed that as the Annual income of Customers is increasing, Monthly Debt is also increasing. From the output, it is clear that Relplot is a figure level function built on top of Facetgrid object.

```python
sns.relplot(x = 'Annual Income', y = 'Monthly Debt',
            data = credit_data, height = 6, aspect = 1.5)
```

As already told,the Relplot function calls underlying axes level function i.e scatterplot

```python
f,ax = plt.subplots(figsize = (12, 8))

sns.scatterplot(x = 'Annual Income', y = 'Current Credit Balance', data = credit_data )
```

Here , We have plotted a scatter plot between 'Number of Credit problems' and 'Tax liens', It can be observed that as the number of tax liens increases for the customers, number of Credit Problems associated with customers also increases. 
The relative opacity of data points can be controlled by 'alpha'. Alpha values can be set between 0 to 1, with high values of alpha means highly opaque data points.Here, alpha is set to low value of 0.1, the data points have low degree of opacity.

```python
sns.relplot(x = 'Tax Liens', y = 'Number of Credit Problems', 
            alpha = 0.1, data = credit_data , aspect = 2)
```

Here, alpha is set to high value 0.8, which means data points are having high degree of opacity as seen from the plot

```python
f,ax = plt.subplots(figsize = (12, 8))

sns.scatterplot(x = 'Tax Liens', y = 'Number of Credit Problems', 
                alpha = 0.6, data = credit_data )
```

In the above fig. showing Scatter plot between 'Monthly Debt' and 'Annual Income',a  Third variable 'Loan Status' is added to the plot by hue,which differentiates levels based on colors. Quite a few outliers are visible in the plot. Most of them have repaid the loan. One of the customer with the highest Annual income has defaulted in loan repayment.

```python
sns.relplot(x = 'Annual Income', y = 'Monthly Debt', hue = 'Loan Status', 
            data = credit_data, height = 6, aspect = 1.5, s=60)
```

 We have added one more variable 'Term' as style where levels of the variable is differentiated based on marker style.
The customer with highest annual income had  taken a Long Term loan and defaulted in loan repayment. Majority of the outliers had taken Short term loan.

```python
sns.relplot( x = 'Annual Income', y = 'Monthly Debt', hue = 'Loan Status', 
            style = 'Term', data = credit_data , height = 6, aspect = 1.5, s=100)
```

We have customised the marker style by using a dictionary specifying Markers for each level of style.

```python
markers = {'Short Term': 'D', 'Long Term': 'P'}

sns.relplot( x = 'Annual Income', y = 'Monthly Debt', hue = 'Loan Status', 
            style = 'Term', markers = markers, data = credit_data, 
            height = 6, aspect = 1.5, s=100)
```

Here we have plotted 'Bankruptcies'vs 'Number of Credit Problems'. It is observed that as the number of credit problems for customers is increasing, the number of Bankruptcies is also increasing. We have added one more variable 'Tax liens' as size and have customised the size ranges. It can be clearly observed that as the number of credit problems for customers is increasing, the number of Tax Liens is also increasing. But there is no such visible realtionship exists between 'Tax liens' and 'Bankruptcies'.

```python
sns.relplot(x = 'Number of Credit Problems', y = 'Bankruptcies', 
            data = credit_data, s = 80,
            height = 6, aspect = 1.5)
```

```python
sns.relplot(x = 'Number of Credit Problems', y = 'Bankruptcies', 
            size = 'Tax Liens', sizes = (20, 400), 
            data = credit_data,
            height = 6, aspect = 1.5)
```

Here , We have plotted 'Annual Income' against 'Monthly Debt'.We have added 'Loan Status' as hue and also added a variable 'Home Ownership' as 'col',which facets through columns. 
It can be observed that quite a large number of outliers are from 'Home mortage' category of 'Home Ownership'.
Outlier with highest Annual income,who had defaulted in loan repayment is from 'Home Mortage' category.
One outlier having a abnormally high 'Monthly Debt' and two ouliers with high 'Annual Income' are from 'Rent' Category of Home Ownership.All of them had repayed the loan.
The data points from 'HaveMortage' category of Home Ownership have both low values of 'Monthly Debt' and 'Annual Income' and data point counts is also low in comparison to other categories.


```python
sns.relplot( x = 'Annual Income', y = 'Monthly Debt', hue = 'Loan Status', s = 100,
             row = 'Home Ownership', data = credit_data , aspect = 1.5);
```

In a similar way, We can add variable through 'row' argument. We have plotted 'Annual Income' against 'Monthly Debt'.We have added 'Loan Status' as hue and also added one more variable 'Purpose' as 'row',which facets through rows. By zooming into the plot, it can be observed that all those customers taking loan for purpose of 'Renewable energy' had not defaulted. Most of customers who had taken loan for Educational purpose had repayed the loan.
The count of data points for 'Debt Consolidation' purpose is very high compared to others.So Majority of customers had taken loan for Debt Consolidation. Quite a few ouliers visible in Debt Consloidation category and Most of them had repaid the loan. Few ouliers are observed under the purpose of  'Home improvements' . Most of them had repaid the loan. Few ouliers are observed under the purpose of 'Business loan' who had defaulted in loan repayment.

```python
sns.relplot( x = 'Annual Income', y = 'Monthly Debt', hue = 'Term', s = 100,
             col = 'Loan Status' , data = credit_data, height = 8);
```
